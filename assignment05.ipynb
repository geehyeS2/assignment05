### import packages for plottion graphs and manipulating data:

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
from skimage import io, color
from skimage import exposure
from math import pi
```

### De􏰀fined derivative in x-direction function with convolution kernel

```python
def der_x(data):
    ker 		= np.array([[1,-1,0],[1,-1,0],[1,-1,0]]) 
    result		= signal.convolve2d(data, ker, boundary='symm', mode='same') 
    return result
```

### De􏰀fined derivative in y-direction function with convolution kernel

```python
def der_y(data):
    ker 		= np.array([[1,-1,0],[1,-1,0],[1,-1,0]]) 
    result		= signal.convolve2d(data.T, ker, boundary='symm', mode='same') 
    return result.T
```

### Defined absolute value of gradient function

```python
def Ab_gradient(x,y):
    gradient_result = ((x**2) + (y**2))**(1/2)
    return gradient_result
```

### Defined direction of gradient function

```python
def gradient_direction(x,y):
    x=np.where(x==0,0.00001, x)
    return np.arctan(y/x)
```

### Defined smoothing function used convolution kernal 

```python
def smooth(data):
    ker 		= np.array([[1/3,1/3,1/3],[1/3,1/3,1/3],[1/3,1/3,1/3]]) 
    result		= signal.convolve2d(data, ker, boundary='symm', mode='same') 
    return result
```

### Defined own function used convolution kernal 

```python
def own(data):
    ker 		= np.array([[1/5,-3/5,1/5],[1/5,-3/5,1/5],[1/5,-3/5,1/5]]) 
    result		= signal.convolve2d(data, ker, boundary='symm', mode='same') 
    return result
```

```python
x=der_x(im_gray)
y=der_y(im_gray)
```
